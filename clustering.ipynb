```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('app_users.csv', index_col = [0])
df.head()
```

```python
sns.set(style='darkgrid', 
        rc={'figure.figsize':(12,8)})

sns.scatterplot(x='visit_per_month', y='use_time', data=df, s=100, palette='bright')
```

```python
model = KMeans(n_clusters=3, random_state=123)
model.fit(df)

# 클러스터 구분하기
df['label'] = model.predict(df)

# 클러스터별 속한 유저 수
df.groupby('label').count()
```

```python
sns.scatterplot(x='visit_per_month', y='use_time', data=df, hue=df['label'], s=150, palette='bright')
```
